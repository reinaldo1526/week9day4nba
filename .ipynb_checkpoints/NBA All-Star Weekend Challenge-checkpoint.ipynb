# What is the average weight of all players who played during this time?

```python

import numpy as np
import pandas as pd

df = pd.read_csv('NBA All Stars 2000-2016 - Sheet1.csv', delimiter=',')
playersw = df.drop_duplicates('Player')
playersw.mean()

# 228.75 lbs.
```

# What team is represented the most during this time?

```python
df.groupby("Team").count().sort_values("Player")

# Miami Heat
```

# What draft class is represeted the most during this time?

```python
df.groupby("NBA Draft Status").count().sort_values("Team")

#1996 Rnd 1 Pick 13
```

# Are foriegn players more prevalent in the All-Star Game during this time?

```python
df.groupby("Nationality").count()

#No. USA
```

# How often are the Western All-Stars voted in by fans?

```python
df.groupby("Selection Type").count()

easternfanvote = df[df['Selection Type'] == 'Eastern All-Star Fan Vote Selection']
westernfanvote = df[df['Selection Type'] == 'Western All-Star Fan Vote Selection']
fanvote = pd.concat([easternfanvote, westernfanvote])
fanvote
fanvote.drop_duplicates("Player")
# Fans voted in 170 players out of a total of 439... about 39% of all players
# 47 players were voted in 170 times
```

# How many times does Steph Curry make the All-Star game during this time period?

# How many Shooting Gaurds have made the All-Star game during this time?

```python
df.groupby("Pos").count()
sgaurds = df[df["Pos"] == "SG"]

sgaurds.groupby("Player").count()
#9 Shooting Guards
```

# How many Chicago Bulls players have made the All-Star Game during this time period?

```python

df.groupby("Team").count()
cbullsplayers = df[df['Team'] == 'Chicago Bulls']

cbullsplayers.groupby("Player").count()

# 5 
```
